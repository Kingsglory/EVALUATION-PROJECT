# How to rename categoriesmusing pandas method

```python
df_new=pd.Categorical(df['Survived'])

df_new= df_new.rename_categories(['Died','Survived'])


df_new.describe()
```

```python
Pclass_New = pd.Categorical(df['Pclass'], ordered=True)


Pclass_New=Pclass_New.rename_categories(['First_class','Second_class','Third_class'])


Pclass_New.describe()
```

```python
New_s=pd.Categorical(df['SibSp'], ordered=True)

New_s=New_s.rename_categories(['T1','T2','T3','T4','T5','T6','T7'])

New_s.describe()
```

```python
#Reassign New name to Pclass


df['Pclass']=Pclass_New
```

```python
#Check if they have changed

df.head(10)
```

```python
df['Cabin'].unique()
```

```python
char_Cabin= df['Cabin'].astype(str)  #conver data to string
#List comprehension (creating list using another set of string, list etc)

New_Cabin = np.array([Cabin[0] for Cabin in char_Cabin]) # Take first letter

New_Cabin = pd.Categorical(New_Cabin)

New_Cabin.describe()
```

```python
df['Cabin']=New_Cabin  #Reassign names to cabin
```

# Find all passengers whose age is missing

```python
## np.where gives the index of those missing data
```

```python
missing = np.where(df['Age'].isnull()==True) #Find all nulls in column Age

missing
```

```python
#Using the index, locate passangers

df.loc[missing]
```

# Find the maximuim fare 

```python
index = np.where(df['Fare'] == max(df['Fare']))

index
```

```python
df.loc[index]
```

```python
inde = np.where(df['Fare'] == min(df['Fare']))

inde
```

```python
df.loc[inde]
```

# Find the family who has highest members?

```python
#Sibsp and parch are siblings and parents/guardian respectively (add both of the and create new column)


df['Family'] = df['SibSp'] + df['Parch']


most_family = np.where(df["Family"] == max(df["Family"]))

df.loc[most_family]
```
