```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a=np.array([1,2,3,4,5,6])
b=np.array([11,12,13,14,15,16])
```

```python
plt.title('linear line')
plt.xlabel('a')
plt.ylabel('b')
plt.plot(a,b)
plt.show()
```

```python
plt.title('linear line')
plt.xlabel('a')
plt.ylabel('b')
plt.plot(a,b,'--bo') #--(=graph line),b=(color),o=(marker ie mark)
plt.show()
```
