# USING ZSCORE TO REMOVE THE OUTLIERS

```python
import numpy as np
import pandas as pd
from scipy.stats import zscore
import warnings
warnings.filterwarnings('ignore')
```

```python
z=np.abs(zscore(df))

z
```

```python
#Taking the threshold to be 3

print(np.where(z>3))
```

```python
df=df[(z<3).all(axis=1)]

df 
```

# The above are the preprossed data using zscore

```python
x = df.drop('Loan_Status', axis = 1)

y = df['Loan_Status']
```

```python
y.value_counts()
```

The above dependent Loan_Status are not balanced.

```python
#Oversampling the data



from imblearn.over_sampling import SMOTE

SM = SMOTE()
x1,y1 = SM.fit_resample(x,y)

```

```python
#To check the value count of target column

y.value_counts()
```

```python
y1.value_counts()
```

The above y1 value_count shows that the data is balanced as they are equal

# MODELLING

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

maxAccu = 0
maxRS = 0

for i in range(1, 200):
    x_train, x_test, y_train, y_test = train_test_split(x1, y1, test_size = 0.30, random_state =i)
    RFR = RandomForestClassifier()
    RFR.fit(x_train, y_train)
    pred = RFR.predict(x_test)
    acc = accuracy_score(y_test, pred)
    if acc>maxAccu:
        maxAccu = acc
        maxRS = i
print('Best accuracy is ', maxAccu, 'at random_state', maxRS)
```

# From the above, the best random state is 91% accuracy at the random_state 8

# CHECKING FOR OTHER BEST ALGORITHM

```python
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier,AdaBoostClassifier

from sklearn.linear_model import LogisticRegression

from sklearn.metrics import classification_report,confusion_matrix, accuracy_score, roc_curve
```

# FOR RANDOMFORESTCLASSIFIER

```python
RFC= RandomForestClassifier()

RFC.fit(x_train, y_train)
```

```python
predRFC=RFC.predict(x_test)
```

```python
print(accuracy_score(y_test,predRFC))

print(confusion_matrix(y_test, predRFC))

print(classification_report(y_test,predRFC ))
```

# FOR LOGISTIC REGRESSION

```python
LR=LogisticRegression()

LR.fit(x_train,y_train)
```

```python
predLR=LR.predict(x_test)
```

```python
print(accuracy_score(y_test,predLR))

print(confusion_matrix(y_test,predLR))

print(classification_report(y_test,predLR))
```

# FOR GRADIENT BOOSTING CLASSIFIER

```python
GB=GradientBoostingClassifier()
GB.fit(x_train,y_train)
```

```python
predGB=GB.predict(x_test)
```

```python
print(accuracy_score(y_test,predGB))

print(confusion_matrix(y_test,predGB))

print(classification_report(y_test,predGB))
```

# FOR ADABOOST CLASSIFIER

```python
AD=AdaBoostClassifier()

AD.fit(x_train,y_train)
```

```python
predAD=AD.predict(x_test)
```

```python
print(accuracy_score(y_test,predAD))

print(confusion_matrix(y_test,predAD))

print(classification_report(y_test, predAD))
```

# FROM THE ABOVE MODEL, THE BEST MODEL WITH HIGHEST ACCURACY IS THE GRADIENTBOOSTINGCLASSIFIER
